# Number guessing game

```python
import random
n=20
guessing_numb=int(n*random.random())+1
guess=0
while guess != guessing_numb:
    guess=int(input("enter number:"))
    if guess>0:
        if guess>guessing_numb:
            print("your number is greater")
        elif guess<guessing_numb:
            print("your number lesser")
    else:
        print("your are giving up")
        break
else:
    print("congrats!!! you gussed it")
```
